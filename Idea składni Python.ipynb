- w Python bardzo często definiujemy własne funkcje, które realizują określone zadania. Poniżej przykład funkcji, która dla zadanej sekwencji wyznacza sekwencję komplementarną.

```python
def komplementarna(x):
    x2 = x.maketrans("ATGC","TACG")
    return x.translate(x2)


##########Przykład użycia
x = "ATTTC"  #przykladowa sekwencja
print(komplementarna(x))
##########
```
